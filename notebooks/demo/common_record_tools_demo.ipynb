## E3 Cadets Users & Host Lookup

```python
lookup = object_lookup.ObjectLookup.load("e3", "cadets", WORK / "provenance-explorer-cache")
cadets_users = lookup.principal_metadata # type: ignore
cadets_hosts = lookup.host_metadata # type: ignore

for _uuid, user in cadets_users.items():
    pprint(user)

print()
for _uuid, host in cadets_hosts.items():
    pprint(host)
```

## E3 Clearscope Common record iterator

```python
start_ns = date_string_to_ns_timestamp("2018-04-05 10:00:00")
end_ns = date_string_to_ns_timestamp("2018-04-05 10:30:00")

logger = drop_log.DropLog()
iterator = iterate_common_records("e3", "clearscope", drop_log=logger, t_start=start_ns, t_end=end_ns)

for i, record in enumerate(iterator):
    pprint(record)
    if (i+1) % 5 == 0:
        break

# note that clearscope has a lot of unresolvable objects; 
logger.summary("e3", "clearscope")
```
